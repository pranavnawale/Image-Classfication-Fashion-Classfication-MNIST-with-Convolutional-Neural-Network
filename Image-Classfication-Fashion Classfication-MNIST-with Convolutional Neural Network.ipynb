# Image-Classfication-Fashion Classfication-MNIST-with Convolutional Neural Network

```python
# Importing the Dependencie
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import seaborn as sns 
from matplotlib import pyplot as plt
from keras import Sequential
```

```python
# Load Data
(X_train,y_train),(X_test,y_test)=keras.datasets.fashion_mnist.load_data()
```

```python
# Define the class label
class_labels=["T-shirt/Top","Trouser",'Pullover',"Dress","Coat","Sandal","Shirt","Sneaker","Bag","Ankel boot"]

'''
0=>T-shirt/Top
1=>Trouser
3=>Pullover
4=>Dress
5=>Coat
6=>Sandal
7=>Shirt
8=>Sneaker
9=>Ankel boot '''
```

### Feature Scaling

```python
X_train=X_train/255
X_test=X_test/255
```

```python
X_train[0]
```

## Build Neural Network

```python
model=keras.models.Sequential([
                         keras.layers.Flatten(input_shape=[28,28]),
                         keras.layers.Dense(units=32,activation='relu'),
                         keras.layers.Dense(units=10,activation='softmax')
])
```

```python
# Summery of model
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=1)
```

```python
model.fit(X_train,y_train,epochs=10)
```

#### Training Model Accuracy is 89%

## Test and Evaluate Neural Network Model

```python
model.evaluate(X_test,y_test)
```

## Split Dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train,y_train, test_size = 0.2,random_state=2020)
print("shape of X_train= ",X_train.shape)
print("shape of X_val= ",X_val.shape)
print("shape of y_train= ",y_train.shape)
print("shape of y_val= ",y_val.shape)
```

## Building CNN - Convolutional Neural Network

```python
model=keras.models.Sequential([
                         keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding='valid',activation='relu',input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128,activation='relu'),
                         keras.layers.Dense(units=10,activation='softmax'),
    
])
```

```python
# Summery of model
model.summary()
```

```python
# Complie the model
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=512,verbose=1,validation_data=(X_val,y_val))
```

##  Convolutional Neural Network - Building the Complex Model

```python
cnn_model=keras.models.Sequential([
                         keras.layers.Conv2D(filters=32,kernel_size=3,strides=(1,1),padding='valid',activation='relu',input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                        
                         keras.layers.Conv2D(filters=32,kernel_size=3,strides=(2,2),padding='same',activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128,activation='relu'),
                         keras.layers.Dropout(0.25),
                         keras.layers.Dense(units=256,activation='relu'),
                         keras.layers.Dropout(0.25),

                         
                         keras.layers.Dense(units=256,activation='relu'),
                         keras.layers.Dense(units=10,activation='softmax'),
    
])
```

```python
# Summery of model
cnn_model.summary()
```

```python
# complie the model
cnn_model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
# Train model
cnn_model.fit(X_train,y_train,epochs=20,batch_size=512,verbose=1,validation_data=(X_val,y_val))
```

```python
cnn_model.evaluate(X_test,y_test)
```
